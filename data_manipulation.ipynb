````python
# List of columns that might have commas or other non-numeric issues
month_columns = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
                 'August', 'September', 'October', 'November', 'December']

for col in month_columns:
    if col in df.columns:
        # Step 1: Ensure the column is treated as string, then remove commas and trim whitespace
        df[col] = df[col].astype(str).str.replace(',', '').str.strip()
        # Step 2: Convert to numeric, coercing any remaining errors to NaN
        df[col] = pd.to_numeric(df[col], errors='coerce')
        # Step 3: Fill NaN values with 0 (or your preferred imputation)
        df[col] = df[col].fillna(0)
        # Step 4: Convert to integer type
        df[col] = df[col].astype(int)
````

````python
# Which Model had the highest total Sales in the dataset?
highest_total_sale_car_model = df1.groupby('Model')['Sales'].sum().sort_values(ascending=False).head(1)
highest_total_sale_car_model
````

````python
# What are the top 5 Makes by total Sales?
top5_car_maker_total_sales= df1.groupby('Make')['Sales'].sum().round(2).sort_values(ascending=False).head(5)
top5_car_maker_total_sales
````

````python
# Which Model has the highest 4year_resale_value?
highest_4years_resale_value_car_model= df1.groupby('Model')['4year_resale_value'].sum().round(2).sort_values(ascending=False).head(1)
highest_4years_resale_value_car_model
````

````python
# Is there a correlation between Sales and 4year_resale_value?
correlation_of_sales_4year_resale = df1['Sales'].corr(df1['4year_resale_value'])
correlation_of_sales_4year_resale
````

````python
# What is the distribution of Vehicle_type (e.g., Passenger, Car) in the dataset? Which type has more models?
vehicle_distribution = df1['Vehicle_type'].value_counts()
print(vehicle_distribution)
# Which type has more model
vehicle_distribution.idxmax()
````

````python
# What is the average Sales figure for each Vehicle_type?
average_sales_by_vehicle_type= df1.groupby('Vehicle_type')['Sales'].mean().round(2)

average_sales_by_vehicle_type
````

````python
# What is the average Horsepower for cars from each Make?
#(Note: This would require converting the Horsepower column to a numeric type first, as it currently contains non-numeric entries).
average_horspower_by_make = df1.groupby('Make')['Horsepower'].mean().sort_values(ascending=False).round(2)
average_horspower_by_make
````

````python
# What is the average Price_thousands for Passenger vehicles compared to Car vehicles?
#(Note: This would also require cleaning and converting the Price_thousands column.)
average_price_of_passenger_vehicle= df1.groupby('Vehicle_type')['Price_thousands'].mean()
average_price_of_passenger_vehicle
````

````python
# Which Model has the best Fuel_efficiency? (Note: This column also needs to be converted to a numeric type).
best_fuel_efficiency_model= df1.groupby('Model')['Fuel_efficiency'].sum().sort_values(ascending=False).head(1)
best_fuel_efficiency_model
print(f'Best Fuel efficiency car model : {best_fuel_efficiency_model}')
````
